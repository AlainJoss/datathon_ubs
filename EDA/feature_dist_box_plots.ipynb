```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../clean_data/cleaned_data.csv')
```

```python
import matplotlib.pyplot as plt

compset_columns = [
    'compset_group_Beauty & Personal Care',
    'compset_group_Food & Beverages',
    'compset_group_Home Improvement & Appliances',
    'compset_group_Luxury & Premium & Mainstream & Entertainment',
    'compset_group_Retail'
]

num_features = ['followers', 'pictures', 'videos', 'comments', 'likes']

for compset in compset_columns:
    df_subset = df[df[compset] == 1]
    if not df_subset.empty:
        fig, axes = plt.subplots(nrows=1, ncols=len(num_features), figsize=(20, 4))
        for i, metric in enumerate(num_features):
            axes[i].boxplot(df_subset[metric], vert=True)
            axes[i].set_yscale('log')  # Applying logarithmic scale
            axes[i].set_title(f'{metric}')
            axes[i].set_xlabel('')

        plt.suptitle(f'Boxplots of Engagement Metrics for {compset}', fontsize=16)
        plt.tight_layout(pad=2)
        plt.show()
```
