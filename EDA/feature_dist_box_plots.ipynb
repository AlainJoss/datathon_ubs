```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../clean_data/data_completed_model.csv')
df.columns 
```

```python
import matplotlib.pyplot as plt

num_features = [
    'likes_per_content',
    'likes_per_content_weekly_change',
    'followers_weekly_change',
    'comments_per_likes_weekly_change',
]

# Creating a figure with subplots
fig, axes = plt.subplots(nrows=1, ncols=len(num_features), figsize=(20, 4))

for i, metric in enumerate(num_features):
    axes[i].boxplot(df[metric].dropna(), vert=True)  # Ensure to drop NaN values to avoid errors
    axes[i].set_title(f'{metric}')
    axes[i].set_xlabel('')

plt.suptitle('Boxplots of Engagement Metrics', fontsize=16)
plt.tight_layout(pad=2)
plt.show()
```
